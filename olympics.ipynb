```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import plotly.express as px
```

Extract Olympics data

```python
# Initialize an empty list to store DataFrames for each year
dfs = []
```

## 1896 Olympics

```python
# Get the URL of the webpage containing the table
olymp1896 = 'https://www.olympedia.org/results/56017'
page1896 = requests.get(olymp1896)
page1896.encoding = 'utf-8'  # Set the encoding of the webpage to 'utf-8'

# Parse the webpage content using BeautifulSoup
soup1896 = BeautifulSoup(page1896.text, 'html.parser')

# Find the table with the class 'table table-striped' on the webpage
table1896 = soup1896.find("table", class_="table table-striped")

# Get the titles (headers) of the table
table1896_titles = table1896.find_all('th')
table1896_titles = [titles.text.strip() for titles in table1896_titles]

# Create a DataFrame with the first 5 column titles as headers
df1896 = pd.DataFrame(columns=table1896_titles[0:5])

# Get all the rows of the table
column_data = table1896.find_all('tr')

# Iterate over each row of the table (excluding the header row)
for row in column_data[1:]:
    # Get all the data cells in the row
    row_data = row.find_all('td')
    # Extract and strip the text from the first 5 cells
    individual_row_data = [data.text.strip() for data in row_data][0:5]

    # Ensure the row has the correct number of columns
    if len(individual_row_data) == len(df1896.columns):
        length = len(df1896)  # Get the current length of the DataFrame
        df1896.loc[length] = individual_row_data  # Add the row to the DataFrame
        
# Add a new column for the Olympics year 1896
df1896['Year'] = 1896

# Append the DataFrame for 1896 to the list of DataFrames
dfs.append(df1896)

# Now dfs list contains the DataFrame for 1896, and can be concatenated later with additional years.

# Display the DataFrame
df1896
```
